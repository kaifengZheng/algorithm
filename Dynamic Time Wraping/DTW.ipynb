```python
import pandas as  pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib widget
```

```python
std_data = pd.read_csv("aufoildemo.nor",delimiter=r" +",skiprows=9,names=["energy", "mu"],engine='python')
computed_data = pd.read_csv("xmu.dat",delimiter=r" +",skiprows=17,names=["omega","e","k","mu","mu0","chi"],engine="python") 
cutoff_data = std_data[std_data["energy"]>=11910]
XANES=cutoff_data[cutoff_data["energy"]<=12000]
plt.plot(XANES["energy"],XANES["mu"])
plt.plot(computed_data["omega"],computed_data["mu"])
```

```python

def dtw(s,t):
    n,m = len(s),len(t)
    dtw_dis = np.zeros((n+1,m+1))
    for i in range(n+1):
        for j in range(m+1):
            dtw_dis[i,j]=np.inf
    dtw_dis[0,0] = 0
    for i in range(1,n+1):
        for j in range(1,m+1):
            cost = abs(s[i-1]-t[j-1])
            dtw_dis[i,j]=cost+np.min([dtw_dis[i-1,j],dtw_dis[i-1,j-1],dtw_dis[i,j-1]])
    return dtw_dis
    
s=np.array(XANES["mu"])
t=np.array(computed_data["mu"])
distance=dtw(s,t)
plt.imshow(distance)
plt.colorbar()
print(distance[-1,-1])
```
